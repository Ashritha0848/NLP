```python
D1  = "I like machine learning"
D2  = "I like deep learning"
D3  = "I like playing tennis"
D4  = "I like playing cricket"
D5  = "Machine learning is very interesting"
D6  = "Deep learning uses neural networks"
D7  = "Neural networks are powerful models"
D8  = "I enjoy watching cricket matches"
D9  = "Tennis is a popular sport"
D10 = "Cricket is played in many countries"
D11 = "Artificial intelligence is the future"
D12 = "AI and machine learning are related fields"
D13 = "Data science includes machine learning"
D14 = "Python is popular for machine learning"
D15 = "Football is a team sport"
D16 = "Basketball is played worldwide"
D17 = "I love programming in Python"
D18 = "Coding improves problem solving skills"
D19 = "Deep learning requires large datasets"
D20 = "Supervised learning uses labeled data"
D21 = "Unsupervised learning finds hidden patterns"
D22 = "Reinforcement learning uses rewards"
D23 = "Sports improve physical fitness"
D24 = "Regular exercise keeps the body healthy"
D25 = "Technology is evolving rapidly"
D26 = "Cloud computing supports big data"
D27 = "Big data analytics is important"
D28 = "Natural language processing is part of AI"
D29 = "Chatbots use natural language processing"
D30 = "Tennis and cricket are outdoor sports"
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

documents = [D1, D2, D3, D4,D5,D6,D7,D8,D9,D10,D11,D12,D13,D14,D15,D16,D17,D18,D19,D20,D21,D22,D23,D24,D25,D26,D27,D28,D29,D30]

# Initialize TfidfVectorizer
vectorizer = CountVectorizer(max_df=0.95, min_df=1, stop_words="english")

# Fit and transform the documents
bow = vectorizer.fit_transform(documents)

# Get feature names (words)
feature_names = vectorizer.get_feature_names_out()

# Convert to DataFrame for better readability
df_bow = pd.DataFrame(bow.toarray(), columns=feature_names)

print("Bag of Words (TF-IDF) Representation:")
print(df_bow)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# Calculate cosine similarity between documents
cosine_sim_matrix = cosine_similarity(df_bow)

# Convert to DataFrame for better readability
df_cosine_sim = pd.DataFrame(cosine_sim_matrix, index=[D1, D2, D3, D4,D5,D6,D7,D8,D9,D10,D11,D12,D13,D14,D15,D16,D17,D18,D19,D20,D21,D22,D23,D24,D25,D26,D27,D28,D29,D30], columns=[D1, D2, D3, D4,D5,D6,D7,D8,D9,D10,D11,D12,D13,D14,D15,D16,D17,D18,D19,D20,D21,D22,D23,D24,D25,D26,D27,D28,D29,D30])

print("Cosine Similarity Matrix:")
print(df_cosine_sim)
```

```python
from sklearn.metrics import jaccard_score
import numpy as np
import pandas as pd

# Convert BOW matrix to binary (presence/absence) for Jaccard similarity
binary_bow = (df_bow > 0).astype(int)

# Calculate Jaccard Similarity
num_documents = binary_bow.shape[0]
jaccard_sim_matrix = np.zeros((num_documents, num_documents))

for i in range(num_documents):
    for j in range(num_documents):
        # For Jaccard similarity, we can use the jaccard_score which computes 1 - Jaccard distance
        # However, jaccard_score expects 1D arrays for binary classification tasks.
        # To compute similarity between two sets represented as binary vectors,
        # we can calculate (intersection_size / union_size)

        # Get the binary vectors for documents i and j
        vec_i = binary_bow.iloc[i]
        vec_j = binary_bow.iloc[j]

        # Calculate intersection and union
        intersection = np.sum(np.logical_and(vec_i, vec_j))
        union = np.sum(np.logical_or(vec_i, vec_j))

        if union == 0:
            jaccard_sim_matrix[i, j] = 0.0 # If both sets are empty, similarity is 0
        else:
            jaccard_sim_matrix[i, j] = intersection / union

# Convert to DataFrame for better readability
df_jaccard_sim = pd.DataFrame(jaccard_sim_matrix, index=documents, columns=documents)

print("Jaccard Similarity Matrix:")
print(df_jaccard_sim)
```

```python

import nltk
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize
import numpy as np
import pandas as pd

try:
    nltk.data.find('corpora/wordnet')
except LookupError:
    nltk.download('wordnet')
try:
    nltk.data.find('tokenizers/punkt_tab')
except LookupError:
    nltk.download('punkt_tab')

D1  = "I like machine learning"
D2  = "I like deep learning"
D3  = "I like playing tennis"
D4  = "I like playing cricket"
D5  = "Machine learning is very interesting"
D6  = "Deep learning uses neural networks"
D7  = "Neural networks are powerful models"
D8  = "I enjoy watching cricket matches"
D9  = "Tennis is a popular sport"
D10 = "Cricket is played in many countries"
D11 = "Artificial intelligence is the future"
D12 = "AI and machine learning are related fields"
D13 = "Data science includes machine learning"
D14 = "Python is popular for machine learning"
D15 = "Football is a team sport"
D16 = "Basketball is played worldwide"
D17 = "I love programming in Python"
D18 = "Coding improves problem solving skills"
D19 = "Deep learning requires large datasets"
D20 = "Supervised learning uses labeled data"
D21 = "Unsupervised learning finds hidden patterns"
D22 = "Reinforcement learning uses rewards"
D23 = "Sports improve physical fitness"
D24 = "Regular exercise keeps the body healthy"
D25 = "Technology is evolving rapidly"
D26 = "Cloud computing supports big data"
D27 = "Big data analytics is important"
D28 = "Natural language processing is part of AI"
D29 = "Chatbots use natural language processing"
D30 = "Tennis and cricket are outdoor sports"

documents = [
    D1,D2,D3,D4,D5,D6,D7,D8,D9,D10,
    D11,D12,D13,D14,D15,D16,D17,D18,D19,D20,
    D21,D22,D23,D24,D25,D26,D27,D28,D29,D30
]

# -------------------------
# WORDNET SIMILARITY FUNCTION
# -------------------------

def document_wordnet_similarity(doc1, doc2):
    tokens1 = word_tokenize(doc1.lower())
    tokens2 = word_tokenize(doc2.lower())

    synsets1 = [s for token in tokens1 for s in wordnet.synsets(token)]
    synsets2 = [s for token in tokens2 for s in wordnet.synsets(token)]

    if not synsets1 or not synsets2:
        return 0.0

    max_similarities = []

    for s1 in synsets1:
        max_sim_for_s1 = 0.0
        for s2 in synsets2:
            sim = s1.path_similarity(s2)
            if sim is not None and sim > max_sim_for_s1:
                max_sim_for_s1 = sim
        max_similarities.append(max_sim_for_s1)

    if max_similarities:
        return np.mean(max_similarities)
    else:
        return 0.0

# -------------------------
# CREATE SIMILARITY MATRIX
# -------------------------

num_documents = len(documents)
wordnet_sim_matrix = np.zeros((num_documents, num_documents))

for i in range(num_documents):
    for j in range(num_documents):
        if i == j:
            wordnet_sim_matrix[i, j] = 1.0
        else:
            wordnet_sim_matrix[i, j] = document_wordnet_similarity(documents[i], documents[j])

df_wordnet_sim = pd.DataFrame(
    wordnet_sim_matrix,
    index=[f"D{i+1}" for i in range(num_documents)],
    columns=[f"D{i+1}" for i in range(num_documents)]
)

print("WordNet Similarity Matrix (Path Similarity):")
print(df_wordnet_sim)
```
